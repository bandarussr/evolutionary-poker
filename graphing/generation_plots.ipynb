## Line Plot: Fitness vs Generation
This plot shows how the population's fitness changes over generations.

```python
labels = [
    "fitness",
    "avg_rounds_lasted",
    "avg_trait_aggressiveness",
    "avg_trait_risk_tolerance",
    "avg_trait_bluff_tendency",
    "avg_trait_adaptability",
    "avg_trait_position_awareness",
    "avg_trait_chip_size_awareness",
    "avg_action_bluff_attempts",
    "avg_action_fold",
    "avg_action_raise",
    "avg_action_call",
    "avg_action_all_in",
    "avg_action_check"
]
```

```python
def plot_label(label):
    plt.figure(figsize=(10, 6))
    plt.plot(population_stat['generation'], population_stat[label])
    plt.xlabel('Generation')
    plt.ylabel(label)
    plt.title(f'{label} over Generations')
    plt.grid(True)
    plt.show()
```

```python
for label in labels:
    plot_label(label)
```
